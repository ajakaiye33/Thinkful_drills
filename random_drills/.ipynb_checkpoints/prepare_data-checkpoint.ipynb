```python
drop_col = ['city','violent_crime', 'rape1','rape2','aggravated_assault','burglary','larceny_theft','motor_vehicle_theft',\
            'arson']
```

```python
crime_data.drop(drop_col, axis=1, inplace = True)
```

```python
crime_data.fillna(method='ffill', inplace= True)
```

```python
crime_data = crime_data.astype(np.int64)
```

```python
crime_data['population2'] = np.square(crime_data['population'])
```

```python
crime_data.head()
```

```python
this_order = ['population','population2','murder', 'robbery', 'property_crime']
crime_data = crime_data[this_order]
```

```python
crime_data['murder'] = np.where(crime_data['murder'] == 0,0,1)
crime_data['robbery'] = np.where(crime_data['robbery']== 0,0,1)
```
